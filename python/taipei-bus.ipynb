```python
driver = webdriver.Chrome("D:\Python training\chromedriver.exe")
driver.get("https://ebus.gov.taipei/")
time.sleep(2)
```

```python
driver.find_element_by_xpath("//input[@id='inputKeyword']").click()
time.sleep(0.5)
q = driver.find_element_by_xpath("//input[@id='inputKeyword']")
q.send_keys("2")
time.sleep(0.8)

q.send_keys(Keys.DOWN)
time.sleep(0.5)
q.send_keys(Keys.UP)
time.sleep(0.5)
q.send_keys(Keys.ENTER)
time.sleep(1)
driver.find_element_by_xpath("//*[@id='findPanel']/div[1]/div/a[3]").click()
```

```python
info_data = driver.find_elements_by_xpath("//ul[@class='stationlist-move-route-bus-info']/li/span[2]")
for data in info_data:
    print(data.text, "\n")
dsad = []
dsad.append(info_data[2].text)
print(dsad)
```

```python
col = []
from_to_data = []
head_end_data = []
headway_data = []
buffer_data = []
charge_data = []
tel_data = []

for (step, route) in enumerate(route_list):
    driver.find_element_by_xpath("//input[@id='inputKeyword']").click()
    time.sleep(1)
    q = driver.find_element_by_xpath("//input[@id='inputKeyword']")
    q.send_keys(route)
    time.sleep(1)
    q.send_keys(Keys.ENTER)

    time.sleep(1)
    driver.find_element_by_xpath("//*[@id='findPanel']/div[1]/div/a[3]").click()
    
    # info_col = driver.find_elements_by_xpath("//ul[@class='stationlist-move-route-bus-info']/li/span[1]")
    info_data = driver.find_elements_by_xpath("//ul[@class='stationlist-move-route-bus-info']/li/span[2]")

    from_to_data.append(info_data[0].text)
    head_end_data.append(info_data[1].text)
    headway_data.append(info_data[2].text)
    buffer_data.append(info_data[3].text)
    charge_data.append(info_data[4].text)
    tel_data.append(info_data[5].text)

    # 返回
    driver.find_element_by_xpath("//img[@src='/images/nav_011.svg']").click()
    # if step == 2:
    #     break
col = ["起迄站名", "頭末班車", "發車間距", "分段緩衝", "收費方式", "業者服務電話"]
df = pd.DataFrame(dict(zip(col, [from_to_data, head_end_data,\
         headway_data, buffer_data, charge_data, tel_data])))
driver.close()
df.to_csv("./taipei-bus.csv", encoding="utf-8", index=False)
df
```
